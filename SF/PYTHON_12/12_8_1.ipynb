```python
import pandas as pd

orders = pd.read_csv('data/orders.csv', sep=';')
products = pd.read_csv('data/products.csv', sep=';')
```

```python
orders.head()
```

```python
products.head()
```

```python
products.rename(columns={'Product_ID' : 'ID товара'}, inplace=True)
products.head()
orders_products = orders.merge(products, how='left', on='ID товара')
orders_products.head(25)
```

```python
orders_products['Итого'] = orders_products[orders_products['Оплачен'] == 'Да']['Price']*orders_products['Количество']
pokup = orders_products.groupby('ID Покупателя')['Итого'].sum()
pokup.head()
```
