```python
import pandas as pd

melb_data = pd.read_csv('data/melb_data_ps.csv', sep=',')
melb_data.head()
```

```python
melb_df = melb_data.drop(['index', 'Coordinates'], axis=1)
melb_df.head()
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
display(melb_df['Date'])
```

```python
melb_df['MonthSale'] = melb_df['Date'].dt.month
melb_df['MonthSale'].value_counts(normalize=True)
```

```python
melb_df['AgeBuilding'] = melb_df['Date'].dt.year - melb_df['YearBuilt']
display(melb_df['AgeBuilding'])
```

```python
melb_df['WeekdaySale'] = melb_df['Date'].dt.dayofweek

weekend_count = melb_df[(melb_df['WeekdaySale'] == 5) | (melb_df['WeekdaySale'] == 6)].shape
display(weekend_count)
```

```python
def get_weekend(weekday):
    if weekday in [5, 6]:return 1
    else: return 0
melb_df['Weekend'] = melb_df['WeekdaySale'].apply(get_weekend)
display(melb_df[melb_df['Weekend'] == 1]['Price'].mean())
```

```python
import pandas as pd
nlo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv', sep=',')
nlo['Time'] = pd.to_datetime(nlo['Time'])
nlo_1 = nlo[nlo['State'] == 'NV']
display(nlo_1.head())
```
